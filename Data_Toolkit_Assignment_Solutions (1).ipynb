### Create a 3D scatter plot using Plotly

```python

import plotly.graph_objects as go

x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]
z = [5, 6, 7, 8, 9]

fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=z, 
        colorscale='Viridis',
        opacity=0.8))])

fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'))

fig.show()
```

# Theoretical Questions and Answers

### What is NumPy, and why is it widely used in Python
NumPy is a powerful Python library used for numerical computing. It provides support for arrays, matrices, and many mathematical functions. It is widely used for its performance, ease of use, and ability to integrate with other libraries like Pandas and SciPy.

### How does broadcasting work in NumPy
Broadcasting allows NumPy to perform operations on arrays of different shapes by expanding the smaller array along the mismatched dimensions automatically.

### What is a Pandas DataFrame
A Pandas DataFrame is a 2-dimensional, size-mutable, and heterogeneous data structure with labeled axes (rows and columns), similar to a table in SQL or Excel.

### Explain the use of the groupby() method in Pandas
The `groupby()` method in Pandas is used to split the data into groups based on some criteria, apply a function (like sum, mean), and then combine the results.

### Why is Seaborn preferred for statistical visualizations
Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics with fewer lines of code.

### What are the differences between NumPy arrays and Python lists
NumPy arrays are more efficient for numerical operations, support vectorized operations, and consume less memory. Python lists are more general and flexible but slower for large data.

### What is a heatmap, and when should it be used
A heatmap is a graphical representation of data where individual values are represented by color. It is commonly used to visualize correlation matrices and patterns in 2D data.

### What does the term vectorized operation mean in NumPy
Vectorized operations allow operations to be applied to entire arrays without explicit loops, leading to cleaner, faster, and more efficient code.

### How does Matplotlib differ from Plotly
Matplotlib is a static, low-level plotting library ideal for publication-quality plots, while Plotly provides interactive, web-based plots with built-in zoom and hover features.

### What is the significance of hierarchical indexing in Pandas
Hierarchical indexing allows for multiple levels of indexing in a DataFrame or Series, enabling more complex data representations and advanced data selection techniques.

### What is the role of Seaborns pairplot() function
The `pairplot()` function in Seaborn plots pairwise relationships in a dataset, helping to understand the distribution and relationships between variables.

### What is the purpose of the describe() function in Pandas
The `describe()` function provides summary statistics (like mean, std, min, max) for numerical columns in a DataFrame.

### Why is handling missing data important in Pandas
Missing data can lead to incorrect analysis or errors in computations. Pandas provides tools to detect, fill, or drop missing values to ensure data integrity.

### What are the benefits of using Plotly for data visualization
Plotly offers interactive plots, web integration, and supports a wide variety of chart types. It is especially useful for dashboards and visual exploration.

### How does NumPy handle multidimensional arrays
NumPy supports n-dimensional arrays through the ndarray object, enabling advanced indexing, slicing, and broadcasting across multiple dimensions.

### What is the role of Bokeh in data visualization
Bokeh is a Python library for creating interactive and real-time visualizations for modern web browsers. It is suitable for streaming and large datasets.

### Explain the difference between apply() and map() in Pandas
`map()` is used for element-wise transformations on Series, while `apply()` can be used on both Series and DataFrames to apply functions row-wise or column-wise.

### What are some advanced features of NumPy
Advanced features of NumPy include broadcasting, universal functions (ufuncs), masked arrays, and integration with C/C++ for performance.

### How does Pandas simplify time series analysis
Pandas provides powerful time series tools such as date parsing, frequency conversion, resampling, and moving window statistics.

### What is the role of a pivot table in Pandas
Pivot tables summarize data by grouping and aggregating. They allow data restructuring and quick insights from large datasets.

### Why is NumPys array slicing faster than Pythons list slicing
NumPy uses contiguous memory blocks and C-level optimizations for slicing, while Python lists are objects with pointers, making slicing slower.

### What are some common use cases for Seaborn?
Seaborn is commonly used for correlation plots, distributions (histogram, KDE), categorical plots (box, violin), and visualizing statistical relationships.
